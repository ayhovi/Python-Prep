6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
var6=5
n=1
fac=1
if var6>0:
    while (n<var6):
        n=n+1
        fac=fac*n
    print(f'{var6} es mayor a 0 y su factorial es:{fac}')    
else:
    print(f'{var6} es menor a 0')
```

9) Imprimir los números primos existentes entre 0 y 30

```python
# Imprimir los números primos entre 0 y 30
for num in range(31):
    if num > 1:
        es_primo = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                es_primo = False
        if es_primo:
            print(num)
```

10) ¿Se puede mejorar el proceso del punto 9? Utilizar las sentencias break y/ó continue para tal fin

```python
# Imprimir los números primos entre 0 y 30
for num in range(31):
    if num > 1:
        es_primo = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                es_primo = False
                break
        if es_primo:
            print(num)
```

11) En los puntos 9 y 10, se diseño un código que encuentra números primos y además se lo optimizó. ¿Es posible saber en qué medida se optimizó?

```python
import timeit

def verificar_primos_con_break():
    for num in range(31):
        if num > 1:
            es_primo = True
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    es_primo = False
                    break
            if es_primo:
                print(num)

def verificar_primos_sin_break():
    for num in range(31):
        if num > 1:
            es_primo = True
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    es_primo = False
            if es_primo:
                print(num)

tiempo_con_break = timeit.timeit(verificar_primos_con_break, number=10000)
tiempo_sin_break = timeit.timeit(verificar_primos_sin_break, number=10000)

print(f"Tiempo con break: {tiempo_con_break:.6f} segundos")
print(f"Tiempo sin break: {tiempo_sin_break:.6f} segundos")
```

12) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
n = 100
while n < 300:
    n += 1
    if n % 12 != 0:
        continue
    print(n)
```

14) Crear un ciclo while que encuentre dentro del rango de 100 a 300 el primer número divisible por 3 y además múltiplo de 6

```python
n = 100
while n < 300:
    if n % 3 == 0 and n % 6 == 0:
        print(n)
        break  # Detenemos el bucle cuando encontramos el primer número
    n += 1
```
